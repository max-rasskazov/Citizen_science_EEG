```python
def metric(y_true, y_pred):
    l2ratio = np.sum((y_true - y_pred)**2) / np.sum(y_true**2)
    return np.sqrt(l2ratio) * 100
```

```python
def load_data(train_eeg_dir="/home/maxrasskazov/data/train/"):
    global train_eeg_names
    train_eeg_names = [x for x in sorted(os.listdir(train_eeg_dir)) 
                     if x[-3:] == ".h5"]
    data = []

    for i, eeg_name in enumerate(train_eeg_names):
    #     print(i, eeg_name)
        try:
            h5_file = h5py.File(train_eeg_dir + eeg_name, 'r')
            a_group_key = list(h5_file.keys())[0]
            eeg_data = np.array(h5_file[a_group_key]).T
            data.append(eeg_data)
        except FileNotFoundError:
            print("can't find {}".format(eeg_name))
    return data
```

```python
def find_bad_channels(file_data):
    ''' finds bad (only zero) channels in a file '''
    n_channels = file_data.shape[1]
    bad_channels = []
    for ch in range(n_channels):
        std = file_data[:, ch].std()
        if (
            std < 1e-10
        ):
            bad_channels.append(ch)
    return bad_channels
```

```python
def transform_data_into_disjoint_samples(file_data, sample_size=SAMPLE_SIZE):
    padded = np.pad(file_data,
                    pad_width=[(0, sample_size - file_data.shape[0] % sample_size), (0,0)],
                    mode='mean')
    samples = padded.reshape(-1, sample_size, file_data.shape[-1])
    return samples
```

```python
for i in range(len(data)):
    data[i] = transform_data_into_disjoint_samples(data[i])
```

```python
from random import sample, seed


def split_test_train(all_samples, holdout_perc=10):
    seed(42)
    all_test = []
    all_train = []
    for file in all_samples:
        test_inds = sample(range(file.shape[0]), file.shape[0] * holdout_perc // 100)
        train_inds = list(set(range(file.shape[0])) - set(test_inds))
        all_test.append(test_inds)
        all_train.append(train_inds)
    return all_train, all_test
```

```python
all_train, all_test = split_test_train(data)
```

```python
from random import choices, choice
seed(42)

def make_batch_old(data=data, all_set_inds=all_train, size=BATCH_SIZE):
    ''' assembles batch consisting of fixed number
    of randomly chosen samples from randomly chosen files'''
    batch = []
    file_inds = choices(range(len(data)), k=size)
    for file_ind in file_inds:
        smpl = data[file_ind][choice(all_set_inds[file_ind]),:,:]
        batch.append(smpl.reshape(-1, *smpl.shape))
    batch = np.concatenate(batch, axis=0)
    return batch, file_inds

def make_set(data=data, all_set_inds=all_test):
    dataset = []
    file_inds = []
    for i, file in enumerate(data):
        dataset.append(np.take(file, all_set_inds[i], axis=0))
        file_inds.extend([i for _ in range(len(all_set_inds[i]))])
    dataset = np.concatenate(dataset, axis=0)
    return dataset, file_inds
```

```python
cum_lens = {}


def make_batch(data=data, all_set_inds=all_train, size=BATCH_SIZE):
    ''' assembles batch consisting of fixed number
    of randomly chosen samples from uniform distribution
    over all the samples from the set'''
    if id(all_set_inds) in cum_lens:
        cum_len = cum_lens[id(all_set_inds)]
    else:
        cum_len = [0]
        for i in range(len(all_set_inds)):
            cum_len.append(cum_len[-1] + len(all_set_inds[i]))
            cum_lens[id(all_set_inds)] = cum_len
    assert size <= cum_len[-1]
    samples = sample(range(cum_len[-1]), size)
    batch = []
    file_inds = []
    for sample_num in samples:
        file = 0
        while sample_num >= cum_len[file + 1]:
            file += 1
        file_inds.append(file)
        ind = sample_num - cum_len[file]
        smpl = data[file][ind,:,:]
        batch.append(smpl.reshape(-1, *smpl.shape))
    batch = np.concatenate(batch, axis=0)
    return batch, file_inds
```

```python
def compute_loss(batch, restored_batch, file_inds, all_bad_channels=all_bad_channels, cuda=True):
    loss = 0
    losses = []
    for i, file_ind in enumerate(file_inds):
        orig = batch[i,:,:]
        restored = restored_batch[i,:,:]
        if all_bad_channels[file_ind]:
            good_ch = [x for x in range(NUM_CHANNELS) if x not in all_bad_channels[file_ind]]
            if cuda:
                good_ch = Variable(torch.cuda.LongTensor(good_ch))
            else:
                good_ch = Variable(torch.LongTensor(good_ch))
            orig = torch.index_select(orig, dim=1, index=good_ch)
            restored = torch.index_select(restored, dim=1, index=good_ch)
        losses.append(F.mse_loss(restored, orig))
    overall = sum(losses) / len(losses)
    return overall
```

```python
from IPython.display import clear_output

def train_autoencoder(model, data,
                      train_samples=all_train, test_samples=all_test,
                      optimizer=None, loss_history=[[], []],
                      batch_size=32, num_iter=100000, min_iter=5000, cuda=True,
                      clip=None, make_opt=False, stop_point=0.99,
                      ch_drop_pr=None, eval_period=100):
    if cuda:
        model.cuda()
    else:
        model.cpu()
        
    if make_opt or not optimizer:
        optimizer = torch.optim.Adam(model.parameters(), weight_decay=1e-10)
        
    if loss_history[0] and loss_history[1]:
        test_history_plot = loss_history[1]
        loss_history[1] = loss_history[1][-1::-eval_period][::-1]

    test_set, test_inds = make_set(data, test_samples)
    variance = test_set.var(axis=(0,1)).mean()
    if cuda:
        test_set = Variable(torch.cuda.FloatTensor(test_set), volatile=True)
    else:
        test_set = Variable(torch.FloatTensor(test_set), volatile=True)
    
    mod_name = model.__repr__()[:model.__repr__().find('(')]
    
    min_test_loss = float('inf')
    
    grad_norms = []
    
    for i in range(num_iter):
        batch, file_inds = make_batch(data, train_samples, batch_size)
        if cuda:
            batch = Variable(torch.cuda.FloatTensor(batch))
        else:
            batch = Variable(torch.FloatTensor(batch))
            
        if ch_drop_pr is not None:
            m = nn.Dropout2d(p=ch_drop_pr)
            batch = torch.transpose(m(torch.transpose(batch, 1, 2)), 1, 2) # modeling channel death
            batch = batch * (1 - ch_drop_pr) # compensating for dropout layer scaling

        restored_batch = model(batch)

        # compute loss
        loss = F.mse_loss(restored_batch, batch)
        #loss = compute_loss(batch, restored_batch, file_inds, cuda=cuda)  
        # train with backprop
        loss.backward()
        
        if clip:
            clip_norm = clip * np.median(grad_norms[-300:]) if  grad_norms else 1
            norm = torch.nn.utils.clip_grad_norm(model.parameters(), clip_norm)
            grad_norms.append(norm)

        optimizer.step()

        optimizer.zero_grad()


        loss_history[0].append(loss.data[0])

        period = eval_period
        
        if i % period == 0:
            loss_test = F.mse_loss(model(test_set), test_set).data[0]
            #loss_test = compute_loss(test_set, model(test_set), test_inds, cuda=cuda).data[0]
            history[1].append(loss_test)
            
            # update best model so far
            if loss_test < 0.999 * min_test_loss:
                torch.save(model, f'{mod_name}_{model.hid_to_compressed.out_features}c_\
{model.rnn_enc.hidden_size}n_{model.rnn_enc.num_layers}l{"_dr" if ch_drop_pr else ""}_curbest.pt')
                min_test_loss = loss_test
            
            stop = False                
            if (len(history[0]) > min_iter
                and (np.median(history[0][-5000:]) + np.min(history[1][-5000 // period:])) /
                    (np.median(history[0][-10000:-5000]) +
                     np.min(history[1][-10000 // period: -5000 // period])) > stop_point):
                stop = True # to stop ineffective learning

            clear_output(True)
            
            print(f'{mod_name} model, {model.hid_to_compressed.out_features} compressed \
{model.rnn_enc.hidden_size} hidden units {model.rnn_enc.num_layers} layers{" dr" if ch_drop_pr else ""}')
            
            if i == 0 and len(history[0]) == 1:    
                test_history_plot = [history[1][0]]
            else:
                test_history_plot.extend([history[1][-2] + (history[1][-1] - history[1][-2]) * j / period
                                          for j in range(1,period + 1)])            
            plt.figure(figsize=(10,7))
            plt.plot(loss_history[0],label='train loss')
            plt.plot(test_history_plot,label='test loss')
            plt.yscale('log')
            plt.legend()
            if stop:
                plt.savefig(f'./{mod_name}_{model.hid_to_compressed.out_features}c_\
{model.rnn_enc.hidden_size}n_{model.rnn_enc.num_layers}l{"_dr" if ch_drop_pr else ""}.png',
                           format='png')
            plt.show()
            
            
            plt.plot(loss_history[0][-2*period:], label='loss')
            plt.legend()
            plt.show()
            
            print('train loss =     ', loss.data[0])
            print('test loss =      ', loss_test)
            print('"min" test loss =', min_test_loss)
            print('test variance =  ', variance)
            print('l2 reg lambda =  ', optimizer.param_groups[0]['weight_decay'])
            
            if stop:
                break
    
    savef(grad_norms, f'grad_norms_{model.hid_to_compressed.out_features}c_\
{model.rnn_enc.hidden_size}n_{model.rnn_enc.num_layers}l{"_dr" if ch_drop_pr else ""}')
    history[1] = test_history_plot
    mod_name = model.__repr__()[:model.__repr__().find('(')]
    torch.save(model, f'{mod_name}_{model.hid_to_compressed.out_features}c_\
{model.rnn_enc.hidden_size}n_{model.rnn_enc.num_layers}l{"_dr" if ch_drop_pr else ""}_{len(loss_history[0])}it.pt')
    savef(loss_history, f'history_{mod_name}_{model.hid_to_compressed.out_features}c_\
{model.rnn_enc.hidden_size}n_{model.rnn_enc.num_layers}l{"_dr" if ch_drop_pr else ""}_{len(loss_history[0])}it')
    del batch, test_set
    torch.cuda.empty_cache()
```

```python
def predict(model, data, file_ind):
    file_data = data[file_ind]
    orig_shape = file_data.shape
    if cuda:
        file_data = Variable(torch.cuda.FloatTensor(file_data.reshape((1, *orig_shape))), volatile=True)
        model.cuda()
    else:
        file_data = Variable(torch.FloatTensor(file_data.reshape((1, *orig_shape))), volatile=True)
        model.cpu()
    predicted = model(file_data).cpu().data.numpy().reshape(orig_shape)
    del file_data
    torch.cuda.empty_cache()
    return predicted
    
    
```

```python
def find_metric_on_file(autoencoder, data, file_ind):
    file = data[file_ind]
    bad = all_bad_channels[file_ind]
    predicted = predict(autoencoder, data, file_ind)
    alive_pred = np.delete(predicted, bad, axis=1)
    alive_true = np.delete(file, bad, axis=1)
    return metric(alive_true, alive_pred)
```

# linear dense

```python
class Dummy:
    pass

class Eeg_dense_lin_AE(nn.Module):
    def __init__(self, num_channels=NUM_CHANNELS, compressed_size=1,):
        super(self.__class__, self).__init__()
        # Encoder:
        self.dense_enc = nn.Linear(num_channels, compressed_size)
        # Decoder:
        self.dense_dec = nn.Linear(compressed_size, num_channels)
        # for compatability:
        self.hid_to_compressed, self.rnn_enc = Dummy(), Dummy()
        self.hid_to_compressed.out_features = compressed_size
        self.rnn_enc.hidden_size, self.rnn_enc.num_layers = 0, 0
        
        
    def forward(self, x):
        assert isinstance(x, Variable) and \
                 (isinstance(x.data, torch.FloatTensor) or
                  isinstance(x.data, torch.cuda.FloatTensor))
        compressed = self.dense_enc(x)
        decompressed = self.dense_dec(compressed)
        return decompressed
```

```python
for compressed in (1, 6, 29, 58):  #   2,
    try:
        model = Eeg_dense_lin_AE(num_channels=NUM_CHANNELS, compressed_size=compressed)

        history = [[], []]

        opt = torch.optim.Adam(model.parameters())

        train_autoencoder(model, data, all_train, all_test,
                          opt, history, batch_size=BATCH_SIZE,
                          num_iter=50000, min_iter=20000, cuda=True,
                          stop_point=0.995, ch_drop_pr=0.005, eval_period=500)

        del model
        torch.cuda.empty_cache()
    except Exception as e:
#         print(compressed, '\n', e, file=open('train_errors.txt', 'a'))
        raise e
```

# 3 layer dense with leaky-relu activation aside of compressed

```python
class Dummy:
    pass

class Eeg_dense_lrelu3_AE(nn.Module):
    def __init__(self, num_channels=NUM_CHANNELS, compressed_size=1,):
        super(self.__class__, self).__init__()
        # Encoder:
        self.dense_enc = nn.Linear(num_channels, (num_channels * 2  + compressed_size) // 3)
        self.dense_enc2 = nn.Linear((num_channels * 2  + compressed_size) // 3,
                                    (num_channels + compressed_size * 2) // 3)
        self.dense_enc3 = nn.Linear((num_channels + compressed_size * 2) // 3, compressed_size)
        # Decoder:
        self.dense_dec = nn.Linear(compressed_size, (num_channels + compressed_size * 2) // 3)
        self.dense_dec2 = nn.Linear((num_channels + compressed_size * 2) // 3,
                                    (num_channels * 2  + compressed_size) // 3)
        self.dense_dec3 = nn.Linear((num_channels * 2  + compressed_size) // 3, num_channels)
        # for compatability:
        self.hid_to_compressed, self.rnn_enc = Dummy(), Dummy()
        self.hid_to_compressed.out_features = compressed_size
        self.rnn_enc.hidden_size, self.rnn_enc.num_layers = 0, 0
        
        
    def forward(self, x):
        assert isinstance(x, Variable) and \
                 (isinstance(x.data, torch.FloatTensor) or
                  isinstance(x.data, torch.cuda.FloatTensor))
        enc_layer1 = F.leaky_relu(self.dense_enc(x))
        enc_layer2 = F.leaky_relu(self.dense_enc2(enc_layer1))
        compressed = self.dense_enc3(enc_layer2)
        
        dec_layer1 = F.leaky_relu(self.dense_dec(compressed))
        dec_layer2 = F.leaky_relu(self.dense_dec2(dec_layer1))
        decompressed = self.dense_dec3(dec_layer2)
        return decompressed
```

# 3 layer dense with leaky-relu activation aside of compressed + linear shortcut

```python
class Dummy:
    pass

class Eeg_dense_lrelu3_linshortcut_AE(nn.Module):
    def __init__(self, num_channels=NUM_CHANNELS, compressed_size=1,):
        super(self.__class__, self).__init__()
        # Encoder:
        self.dense_enc = nn.Linear(num_channels, (num_channels  + compressed_size) // 2)
        self.dense_enc2 = nn.Linear((num_channels + compressed_size) // 2,
                                    compressed_size)
        self.dense_enc3 = nn.Linear(num_channels + compressed_size, compressed_size)
        # Decoder:
        self.dense_dec = nn.Linear(compressed_size, (num_channels + compressed_size) // 2)
        self.dense_dec2 = nn.Linear((num_channels + compressed_size) // 2,
                                    num_channels)
        self.dense_dec3 = nn.Linear(num_channels + compressed_size, num_channels)
        # for compatability:
        self.hid_to_compressed, self.rnn_enc = Dummy(), Dummy()
        self.hid_to_compressed.out_features = compressed_size
        self.rnn_enc.hidden_size, self.rnn_enc.num_layers = 0, 0
        
        
    def forward(self, x):
        assert isinstance(x, Variable) and \
                 (isinstance(x.data, torch.FloatTensor) or
                  isinstance(x.data, torch.cuda.FloatTensor))
        enc_layer1 = F.leaky_relu(self.dense_enc(x))
        enc_layer2 = F.leaky_relu(self.dense_enc2(enc_layer1))
        enc_stack = torch.cat([enc_layer2, x], dim=-1)
        compressed = self.dense_enc3(enc_stack)
        
        dec_layer1 = F.leaky_relu(self.dense_dec(compressed))
        dec_layer2 = F.leaky_relu(self.dense_dec2(dec_layer1))
        dec_stack = torch.cat([dec_layer2, compressed], dim=-1)
        decompressed = self.dense_dec3(dec_stack)
        return decompressed
```
