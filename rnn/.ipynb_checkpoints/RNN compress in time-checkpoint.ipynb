```python
def metric(y_true, y_pred):
    l2ratio = np.sum((y_true - y_pred)**2) / np.sum(y_true**2)
    return np.sqrt(l2ratio) * 100
```

```python
def load_data(train_eeg_dir="/home/maxrasskazov/data/train/"):
    train_eeg_names = [x for x in sorted(os.listdir(train_eeg_dir)) 
                     if x[-3:] == ".h5"]
    data = []

    for i, eeg_name in enumerate(train_eeg_names):
    #     print(i, eeg_name)
        try:
            h5_file = h5py.File(train_eeg_dir + eeg_name, 'r')
            a_group_key = list(h5_file.keys())[0]
            eeg_data = np.array(h5_file[a_group_key]).T
            data.append(eeg_data)
        except FileNotFoundError:
            print("can't find {}".format(eeg_name))
    return data
```

```python
def find_bad_channels(file_data):
    n_channels = file_data.shape[1]
    bad_channels = []
    for ch in range(n_channels):
        std = file_data[:, ch].std()
        if (
            std < 1e-10
        ):
            bad_channels.append(ch)
    return bad_channels
```

#### clip and scale data

```python
def scale_clip(data_train, data_test, scaling_factor=1000):
    
    for i in range(len(data_train)):
        data_train[i] *= scaling_factor
        np.clip(data_train[i], a_min=-1, a_max=1, out=data_train[i])
    for i in range(len(data_test)):
        data_test[i] *= scaling_factor
```

transformation for learning

```python
def transform_data_into_disjoint_samples(file_data, sample_size):
    padded = np.pad(file_data,
                pad_width=[(0, sample_size - file_data.shape[0] % sample_size), (0,0)],
                    mode='mean')
    samples = padded.reshape(-1, sample_size, file_data.shape[-1])
    return samples
```

```python
from random import sample, seed


def split_test_train(all_samples, holdout_perc=10):
    seed(42)
    all_test = []
    all_train = []
    for file in all_samples:
        test_inds = sample(range(file.shape[0]), file.shape[0] * holdout_perc // 100)
        train_inds = list(set(range(file.shape[0])) - set(test_inds))
        all_test.append(test_inds)
        all_train.append(train_inds)
    return all_train, all_test
```

```python
from random import choices, choice
seed(42)

# def make_batch_old(data=data, all_set_inds=all_train, size=BATCH_SIZE):
#     ''' assembles batch consisting of fixed number
#     of randomly chosen samples from randomly chosen files'''
#     batch = []
#     file_inds = choices(range(len(data)), k=size)
#     for file_ind in file_inds:
#         smpl = data[file_ind][choice(all_set_inds[file_ind]),:,:]
#         batch.append(smpl.reshape(-1, *smpl.shape))
#     batch = np.concatenate(batch, axis=0)
#     return batch, file_inds

def make_set(data, all_set_inds):
    dataset = []
    file_inds = []
    for i, file in enumerate(data):
        dataset.append(np.take(file, all_set_inds[i], axis=0))
        file_inds.extend([i for _ in range(len(all_set_inds[i]))])
    dataset = np.concatenate(dataset, axis=0)
    return dataset, file_inds
```

```python
cum_lens = {}


def make_batch(data, all_set_inds, size=BATCH_SIZE):
    ''' assembles batch consisting of fixed number
    of randomly chosen samples from uniform distribution
    over all the samples from the set'''
    if id(all_set_inds) in cum_lens:
        cum_len = cum_lens[id(all_set_inds)]
    else:
        cum_len = [0]
        for i in range(len(all_set_inds)):
            cum_len.append(cum_len[-1] + len(all_set_inds[i]))
            cum_lens[id(all_set_inds)] = cum_len
    assert size <= cum_len[-1]
    samples = sample(range(cum_len[-1]), size)
    batch = []
    file_inds = []
    for sample_num in samples:
        file = 0
        while sample_num >= cum_len[file + 1]:
            file += 1
        file_inds.append(file)
        ind = sample_num - cum_len[file]
        smpl = data[file][ind,:,:]
        batch.append(smpl.reshape(-1, *smpl.shape))
    batch = np.concatenate(batch, axis=0)
    return batch, file_inds
```

```python
def compute_loss(batch, restored_batch, file_inds, all_bad_channels, cuda=True):
    loss = 0
    losses = []
    for i, file_ind in enumerate(file_inds):
        orig = batch[i,:,:]
        restored = restored_batch[i,:,:]
        if all_bad_channels[file_ind]:
            good_ch = [x for x in range(NUM_CHANNELS) if x not in all_bad_channels[file_ind]]
            if cuda:
                good_ch = Variable(torch.cuda.LongTensor(good_ch))
            else:
                good_ch = Variable(torch.LongTensor(good_ch))
            orig = torch.index_select(orig, dim=1, index=good_ch)
            restored = torch.index_select(restored, dim=1, index=good_ch)
        losses.append(F.mse_loss(restored, orig))
    overall = sum(losses) / len(losses)
    return overall
```

```python
from IPython.display import clear_output

def train_autoencoder(model, data,
                      train_samples, test_samples,
                      optimizer=None, loss_history=[[], []],
                      batch_size=32, num_iter=100000, min_iter=10000, cuda=True,
                      clip=None, make_opt=False, stop_point=0.99,
                      ch_drop_pr=None, eval_period=100):
    if cuda:
        model.cuda()
    else:
        model.cpu()
        
    if make_opt or not optimizer:
        optimizer = torch.optim.Adam(model.parameters(), weight_decay=0)
        
    if loss_history[0] and loss_history[1]:
        test_history_plot = loss_history[1]
        loss_history[1] = loss_history[1][-1::-eval_period][::-1]

    test_set, test_inds = make_set(data, test_samples)
    variance = test_set.var(axis=(0,1)).mean()
    if cuda:
        test_set = Variable(torch.cuda.FloatTensor(test_set), volatile=True)
    else:
        test_set = Variable(torch.FloatTensor(test_set), volatile=True)
    
    mod_name = model.__repr__()[:model.__repr__().find('(')]
    
    min_test_loss = float('inf')
    
    grad_norms = []
    
    for i in range(num_iter):
        batch, file_inds = make_batch(data, train_samples, batch_size)
        if cuda:
            batch = Variable(torch.cuda.FloatTensor(batch))
        else:
            batch = Variable(torch.FloatTensor(batch))
            
        if ch_drop_pr is not None:
            m = nn.Dropout2d(p=ch_drop_pr)
            batch = torch.transpose(m(torch.transpose(batch, 1, 2)), 1, 2) # modeling channel death
            batch = batch * (1 - ch_drop_pr) # compensating for dropout layer scaling

        restored_batch = model(batch)

        # compute loss
        loss = F.mse_loss(restored_batch, batch)
        #loss = compute_loss(batch, restored_batch, file_inds, cuda=cuda)  
        # train with backprop
        loss.backward()
        
        if clip:
            clip_norm = clip * np.median(grad_norms[-300:]) if  grad_norms else 1
            norm = torch.nn.utils.clip_grad_norm(model.parameters(), clip_norm)
            grad_norms.append(norm)

        optimizer.step()

        optimizer.zero_grad()


        loss_history[0].append(loss.data[0])

        period = eval_period
        
        def model_repr():
            return f'{mod_name}_{round(model.rnn_enc.hidden_size*2/(SAMPLE_SIZE*NUM_CHANNELS), 2)}a_{SAMPLE_SIZE}s_\
{model.rnn_enc.hidden_size}n_{model.rnn_enc.num_layers}l{"_dr" if ch_drop_pr else ""}'
        
        if i % period == 0:
            loss_test = F.mse_loss(model(test_set), test_set).data[0]
            #loss_test = compute_loss(test_set, model(test_set), test_inds, cuda=cuda).data[0]
            history[1].append(loss_test)
            
            # update best model so far
            if loss_test < 0.999 * min_test_loss:
                torch.save(model, model_repr() + '_curbest.pt')
                min_test_loss = loss_test
            
            stop = False                
            if (len(history[0]) > min_iter
                and (np.median(history[0][-(min_iter // 2):]) + np.min(history[1][-(min_iter // 2 // period):])) /
                    (np.median(history[0][-min_iter:-(min_iter // 2)]) +
                     np.min(history[1][-min_iter // period: -(min_iter // 2 // period)])) > stop_point):
                stop = True # to stop ineffective learning

            clear_output(True)
            
            print(model_repr())
            
            if i == 0 and len(history[0]) == 1:    
                test_history_plot = [history[1][0]]
            else:
                test_history_plot.extend([history[1][-2] + (history[1][-1] - history[1][-2]) * j / period
                                          for j in range(1,period + 1)])            
            plt.figure(figsize=(10,7))
            plt.plot(loss_history[0],label='train loss')
            plt.plot(test_history_plot,label='test loss')
            plt.yscale('log')
            plt.legend()
            if stop:
                plt.savefig('./' + model_repr() + '.png', format='png')
            plt.show()
            
            
            plt.plot(loss_history[0][-2*period:], label='loss')
            plt.legend()
            plt.show()
            
            print('train loss =     ', loss.data[0])
            print('test loss =      ', loss_test)
            print('"min" test loss =', min_test_loss)
            print('test variance =  ', variance)
            print('l2 reg lambda =  ', optimizer.param_groups[0]['weight_decay'])
            
            if stop:
                break
    
    savef(grad_norms, 'grad_norms_' + model_repr())
    history[1] = test_history_plot
    torch.save(model, model_repr() + f'_{len(loss_history[0])}it.pt')
    savef(loss_history, model_repr() + f'_{len(loss_history[0])}it')
    del batch, test_set
    torch.cuda.empty_cache()
```

```python
def rsqr(y_true, y_pred):
    return 1 - np.mean((y_true - y_pred)**2) / np.var(y_true)

def mse(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)
```

```python
def predict(model, data_test, file_ind, sample_size):
    file_data = data_test[file_ind]
    file_data = file_data[:file_data.shape[0] // sample_size * sample_size]
    orig_shape = file_data.shape
    set1 = file_data.reshape(-1, sample_size, orig_shape[-1])
    if cuda:
        set1 = Variable(torch.cuda.FloatTensor(set1), volatile=True)
        model.cuda()
    else:
        set1 = Variable(torch.FloatTensor(set1), volatile=True)
        model.cpu()
    pred1 = model(set1).cpu().data.numpy().reshape(orig_shape)
    del set1  #, set2
    torch.cuda.empty_cache()
    return pred1 #+ pred2
    
    
```

```python
def find_metric_on_file(autoencoder, data_test, file_ind, sample_size, all_bad_channels):
    file = data_test[file_ind]
    file = file[:file.shape[0] // sample_size * sample_size]
    bad = all_bad_channels[file_ind]
    predicted = predict(autoencoder, data_test, file_ind, sample_size)
    alive_pred = np.delete(predicted, bad, axis=1)
    alive_true = np.delete(file, bad, axis=1)
    return metric(alive_true, alive_pred), rsqr(alive_true, alive_pred), mse(alive_true, alive_pred)
```

# RNN in time

```python
class Eeg_time_LSTMLoopAE(nn.Module):
    def __init__(self, num_channels=NUM_CHANNELS,
                 rnn_num_units=10, rnn_layers=1):
        super(self.__class__, self).__init__()
        self.rnn_num_units = rnn_num_units
        # Encoder:
        self.rnn_enc = nn.LSTM(num_channels, rnn_num_units,
                              num_layers=rnn_layers, batch_first=True)
#         self.hid_to_compressed = nn.Linear(rnn_num_units, compressed_size)
        # Decoder:
        self.rnn_dec = nn.LSTM(rnn_num_units, rnn_num_units,
                              num_layers=rnn_layers, batch_first=True)
        self.hid_to_decompressed = nn.Linear(rnn_num_units, num_channels)
        
    def forward(self, x):
        assert isinstance(x, Variable) and \
                 (isinstance(x.data, torch.FloatTensor) or
                  isinstance(x.data, torch.cuda.FloatTensor))
        _, (h_seq, c_seq) = self.rnn_enc(x)
        compressed = (h_seq, c_seq)
        
        dummy_size = list(x.size())
        dummy_size[-1] = self.rnn_num_units
        dummy_input = Variable(torch.zeros(*dummy_size)).cuda()
        dec_rnn_h_seq, _ = self.rnn_dec(dummy_input, compressed)
        decompressed = self.hid_to_decompressed(dec_rnn_h_seq)
        return decompressed
```

```python
# pd.DataFrame(columns=['sample_size', 'rnn_units', 'alpha', 'val_loss'])
rnn_time_df = pd.read_csv('rnn_time.csv', index_col=0)
hidden_sizes = {16: [16],
                32: [16, 96]}
'''{2: [2, 29, 58],
                4: [2, 12, 116],
                8: [8, 116],
                16: [8, 48],
                32: [32],
                64: [32, 192],
                128: [128],
                256: [128]}'''

s_sizes = [2**i for i in range(4, 6)] #1, 9
data = data_train = 1
for SAMPLE_SIZE in s_sizes:  # 1,
    del data, data_train
    data = load_data()    
    data_train, data_test = divide_data(data)
    scale_clip(data_train, data_test, scaling_factor=1000)
    del data_test
    
    # epoch in prev tries:
    # 8.6 * 10**6 / (32 * 10**3) ~= 270
    # 5000 ~= 18.6 * epoch
    epoch = sum(x.shape[0] for x in data_train) // (BATCH_SIZE * SAMPLE_SIZE)
    print(epoch)
    
    for i in range(len(data_train)):
        data_train[i] = transform_data_into_disjoint_samples(data_train[i], sample_size=SAMPLE_SIZE)
    all_train, all_val = split_test_train(data_train)
    
    for hid_size in hidden_sizes[SAMPLE_SIZE]:
        
        try:
            model = Eeg_time_LSTMLoopAE(num_channels=NUM_CHANNELS,
                                         rnn_num_units=hid_size, rnn_layers=1)

            history = [[], []]

            opt = torch.optim.Adam(model.parameters())

            train_autoencoder(model, data_train, all_train, all_val,
                              opt, history, batch_size=BATCH_SIZE,
                              num_iter=100 * epoch, min_iter=2 * epoch, cuda=True,
                              stop_point=0.99, clip=2, eval_period=epoch // 10, ch_drop_pr=0.005)
            
            val_set, val_inds = make_set(data_train, all_val)
            val_set = Variable(torch.cuda.FloatTensor(val_set), volatile=True)
            loss_val = F.mse_loss(model(val_set), val_set).data[0]
            rnn_time_df.loc[len(rnn_time_df)] = (SAMPLE_SIZE, hid_size, hid_size*2/(SAMPLE_SIZE*NUM_CHANNELS), loss_val)

            del model, val_set
            torch.cuda.empty_cache()
        except Exception as e:
            raise e
            print(compressed, '\n', e, file=open('train_errors.txt', 'a'))
```

```python
for col in ['train_metric', 'train_r2', 'train_mse', 'test_metric', 'test_r2', 'test_mse']:
    rnn_time_df[col] = None
for (k_tr, train_m), (k_ts, test_m) in zip(train_metrics.items(), test_metrics.items()):
    assert k_tr == k_ts
    train_metric, train_r2, train_mse = [np.mean([x[i] for x in train_m]) for i in range(3)]
    test_metric, test_r2, test_mse = [np.mean([x[i] for j, x in enumerate(test_m) if j not in [12,14,15]])
                                      for i in range(3)]
#     print((train_metric, train_r2, train_mse, test_metric, test_r2, test_mse))
    ind = rnn_time_df[(rnn_time_df.sample_size==k_tr[0]) & (rnn_time_df.rnn_units==k_tr[1])].index
    rnn_time_df.loc[ind,['train_metric', 'train_r2', 'train_mse', 'test_metric', 'test_r2', 'test_mse']
                     ] = (train_metric,   train_r2,   train_mse,   test_metric,   test_r2,   test_mse)
    
```
